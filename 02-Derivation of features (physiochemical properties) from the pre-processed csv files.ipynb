```python
import pandas as pd
import numpy as np
import glob
from pandas import DataFrame
```

# Converting CGC fasta file into dataframe

```python
from Bio import SeqIO
with open(r"data\feature_files\fasta_cgc (4).fasta") as fasta_file:
    identifiers = []
    accession =[]
    for seq_record in SeqIO.parse(fasta_file,'fasta'):
        identifiers.append(str(seq_record.seq))
        accession.append(str(seq_record.id))
```

```python
d={'Entry':accession,'Sequence': identifiers}
data1=pd.DataFrame(d)
data1['Class']='CGC'
data1
```

```python
#Splitting the Entry column to get suffix of the identifiers
data1[['First','Entry']] = data1['Entry'].str.split('_', expand=True)
data1
```

```python
#Dropping prefix from the identifiers
data1.drop('First',axis=1,inplace=True)
```

```python
data1
```

```python
#Exporting the dataframe to a csv file
data1.to_csv('data/feature_files/cgc_foranalysis.csv', index=False)
```

## Get amino acid Column Names from property file

```python
unique_amino_acid = prop.columns.tolist()
```

# Get each amino acid Count from Sequence column in big_Frame file

```python
#data['A'] = list(map(lambda x: x.count("A"), data['Sequence']))
```

```python
for amino_acid in unique_amino_acid:
    big_frame[amino_acid] = list(map(lambda x: x.count(amino_acid), big_frame['Sequence']))
```

# Get Properties Feature

## Get Unique Propery Names

```python
unique_property_names = prop.index.tolist()
unique_property_names
```

## Create one Column

```python
big_frame['sum_K0'] = sum(big_frame[unique_amino_acid].iloc[0] * prop.loc['IK0'])
```

```python
big_frame['IK0'] = sum(big_frame[unique_amino_acid].iloc[0] * prop.loc['IK0']) / big_frame['sequence_length']
```

```python
big_frame.head()
```

```python
big_frame.shape
```

```python
big_frame.drop(['sum_K0','IK0'], axis = 1, inplace=True)
```

## For All Features

```python
for property_name in unique_property_names:
    big_frame[property_name] = sum(big_frame[unique_amino_acid].iloc[0] * prop.loc[property_name]) / big_frame['sequence_length']
    
```

```python
big_frame[unique_amino_acid].shape[0]
```

```python
for property_name in unique_property_names:
    for row in range(big_frame[unique_amino_acid].shape[0]):
        big_frame[property_name] = sum(big_frame[unique_amino_acid].iloc[row] * prop.loc[property_name]) / big_frame['sequence_length']
```
